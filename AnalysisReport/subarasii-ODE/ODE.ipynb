```julia
using DataFrames
using Plots
using DifferentialEquations
using ParameterizedFunctions
using ImageMagick
```

```julia
Base.@kwdef struct LotkaVolterra
    du :: Vector{Float64}
    u  :: Vector{Float64}
    p  :: Vector{Float64}
    t  :: Float64
# end
```

```julia
function lotka_volterra!(du,u,p,t)
    du[1] = - p[1]*u[1]*u[2] + p[2]*u[1]
    du[2] = p[3]*u[1]*u[2] - p[4]*u[2]
end

u0 = [10.0, 7.0]
p = [0.3, 2.0, 0.4, 3.0]
tspan = (0.0, 10.0)
```

```julia
problem = ODEProblem(lotka_volterra!,u0,tspan,p)
```

```julia
sol = solve(problem)
```

```julia
plot(sol)
```

```julia
plot(sol,vars=(1,2))
```

```julia
animate(sol,vars=(1,2),fps=30,lw=2,every=1)
```
